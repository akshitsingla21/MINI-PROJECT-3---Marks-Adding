```python
import pandas as pd

df = pd.read_csv('/content/Mini Project - Marks Adding.csv')
df.head()

csv2dict = pd.read_csv('/content/Mini Project - Marks Adding.csv').to_dict()
print(csv2dict)
k=0

new_csv = {'Roll Num':{}, 'Marks':{}}
L1=[]
k=0
for i in csv2dict['Roll Num']:
  sum=0
  new_csv['Roll Num'][k]=csv2dict['Roll Num'][i]
  new_csv['Marks'][k]=csv2dict['Marks'][i]

  for j in range(i+1, list(csv2dict['Roll Num'].keys())[-1]):
   
     if csv2dict['Roll Num'][i]==csv2dict['Roll Num'][j]:

       L1.append(j)
       sum=sum+ csv2dict['Marks'][j]
  new_csv['Marks'][k]=sum
  k=k+1 
L1.sort()  

L2 = list(set(L1))

print(L2)      
for i in L2:
   del  new_csv['Roll Num'][i] 
   del  new_csv['Marks'][i] 
print( new_csv)      

   
      
      

      
      
  
   
```
